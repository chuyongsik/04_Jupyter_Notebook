```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
plt.plot(df.index,df['출생아 수'], 'r--', label='출생아 수')
plt.plot(df.index,df['합계 출산율'], 'g', label='합계 출산율')
plt.legend()
```

```python
fig, ax1 = plt.subplots(figsize=(10,5))

ax1.bar(df.index,df['출생아 수'], color='#f0700e', label='출생아 수')
ax1.set_ylabel('출생아 수')
ax1.set_ylim(200,650,100) # 출생아 수의 범위를 300~700으로 설정
ax1.set_yticks( np.arange(300,700,100) ) # 출생아 수의 눈금을 300~700까지 100단위로 설정
ax1.grid( linestyle='--', linewidth=1, axis='y', color='g', alpha=0.4)

ax2=ax1.twinx()
ax2.plot(df.index,df['합계 출산율'], color='#f0b40e',marker='o', ms=15, mec='w',mew=3, lw=3)
ax2.set_ylabel('합계 출산율')
ax2.set_ylim(0,1.5)
ax2.set_yticks(np.arange(0,1.5,0.5))

fig.suptitle('출생아 수와 합계 출산율', fontsize=20)

ax1.spines['top'].set_visible(False)
ax2.spines['top'].set_visible(False)
```
