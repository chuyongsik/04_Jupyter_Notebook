#### 데이터 시각화

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
plt.figure(figsize=(10,10))
plt.grid(axis='x', linestyle='--', linewidth=0.5, alpha=1)
plt.barh(df_m.columns, -df_m.iloc[0]/1000) # 남자 전국 인구수 (천명) 오른쪽 그래프
plt.barh(df_w.columns, df_w.iloc[0]/1000) # 여자 전국 인구수 (천명) 왼쪽 그래프
plt.title('2023년 12월 기준 전국의 남녀 인구 분포')
plt.xticks(range(-2500, 2501, 500))
plt.show()
```

```python
df_t = df_m + df_w
pop_death = [949,949,796,796,2757,2757,4427,4427,11539,11539,26341,26341,48998,48998,76621,76621,200493,200493,0,0,0]
pop_death = [int(x/1000/2*10) for x in pop_death]
print(pop_death)

plt.figure(figsize=(10,10))


# Original bars
bars=plt.barh(df_t.columns,  df_t.iloc[0]/1000, color='b') # 남자 전국 인구수 (천명) 오른쪽 그래프

# Reduced bars
bars_reduced = plt.barh(df_t.columns, df_t.iloc[0]/1000- pop_death, color='r')

plt.title('2023년 12월 기준 전국의 남녀 인구 분포')
plt.xticks(range(0, 5500, 370))

# Add data labels for original bars
for bar in bars:
    width = bar.get_width()
    plt.text(width+500, bar.get_y() + bar.get_height()/2, f'{int(width):,}', 
             va='center', ha='left')

# Add data labels for reduced bars
for bar in bars_reduced:
    width = bar.get_width()
    plt.text(width+200, bar.get_y() + bar.get_height()/2, f'{int(width):,}', 
             va='center', ha='left', color='black')

plt.show()
```
