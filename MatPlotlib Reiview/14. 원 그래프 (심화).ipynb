```python
values = [30,25,20,13,10,2]
labels = ['Python','Java','Javascript','C#','C/C++',"etc"]

plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False, explode=(0,0,0.1,0,0,0))
plt.show()
```

```python
# plt.figure(figsize=(10,25))
values = [30,25,20,13,10,2]
labels = ['Python','Java','Javascript','C#','C/C++',"etc"]
explode = [0.05]*6



colors = [(1, 0, 0, 0.5), (0, 1, 0, 0.5), (0, 0, 1, 0.5), (1, 1, 0, 0.5), (1, 0.5, 0, 0.5), (0.5, 0, 0.5)]  # RGBA colors with 0.5 alpha
plt.pie(values, labels=labels, autopct='%.1f%%', startangle=0, counterclock=False, explode=explode, colors=colors)
plt.legend(loc=(1.2,0.3))
plt.show()
```

```python
plt.pie(values, labels=labels, autopct='%.1f%%', startangle=0, counterclock=False, explode=explode, colors=colors,
        wedgeprops={'width': 0.2, 'edgecolor': 'w', 'linewidth': 5})
plt.legend(loc=(1.2,0.3))
```

```python
def custom_autopct(pct):
    # return '{:.1f}%'.format(pct) if pct > 10 else ''
    return '{:.1f}%'.format(pct) if pct > 10 else ''


plt.pie(values, labels=labels, autopct=custom_autopct, startangle=0, counterclock=False, explode=explode, colors=colors,
        wedgeprops={'width': 0.2, 'edgecolor': 'w', 'linewidth': 5})
plt.legend(loc=(1.2,0.3))
```
