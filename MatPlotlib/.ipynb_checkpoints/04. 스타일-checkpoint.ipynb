# 04. 스타일

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='MalGun Gothic'
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus']=False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
x=[1,2,3]
y=[2,4,8]
```

```python
plt.plot(x,y, linewidth=5)
```

## Marker

```python
plt.plot(x,y, marker='o')
```

```python
plt.plot(x,y, marker='v')
```

```python
plt.plot(x,y, marker='v', markersize=10)
```

```python
plt.plot(x,y, marker='X', markersize=10)
```

```python
plt.plot(x,y, marker='v', markersize=10, markeredgecolor='red')
```

```python
plt.plot(x,y, marker='o', markersize=10, markeredgecolor='red')
```

```python
plt.plot(x,y, marker='o', markersize=10, markeredgecolor='red', markerfacecolor='yellow')
```

## 선 스타일

```python
plt.plot(x,y, linestyle=':')
```

```python
plt.plot(x,y, linestyle='dashed')
```

```python
plt.plot(x,y, linestyle='-.')
```

```python
plt.plot(x,y, linestyle='-')
```

## 색깔

```python
plt.plot(x,y, color='#ff0000')
```

```python
plt.plot(x,y, color='g')
```

```python
plt.plot(x,y, color='r')
```

```python
plt.plot(x,y, color='#00ff00')
```

## 포맷


```python
plt.plot(x,y,'ro--')
```

```python
plt.plot(x,y,'bv--')
```

```python
plt.plot(x,y,'go-')
```

## 축약어

```python
plt.plot(x,y, marker='o', mfc='red',ms=10,mec='blue',ls=':')
```

## 투명도

```python
plt.plot(x,y, marker='o', mfc='red',ms=10,mec='blue',ls=':', alpha=0.3)
```

## 그래프 크기

```python
plt.figure(figsize=(10,5))
plt.plot(x,y)
```

```python
plt.figure(figsize=(5,10))
plt.plot(x,y)
```

```python
plt.figure(figsize=(10,5), dpi=100
          )
plt.plot(x,y)
```

## 배경색

```python
plt.figure(facecolor='#0390fc')
plt.plot(x,y)
```
