# 8. 막대 그래프 (기본)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='MalGun Gothic'
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus']=False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
labels=['강백호','서태웅','정대만'] # 라벨
values=[190,187,184] # 키
plt.bar(labels,values) # 막대그래프
plt.title('농구선수 키 비교') # 제목
```

```python
plt.bar(labels,values, color='r') # 막대그래프
```

```python
labels=['강백호','서태웅','정대만'] # 라벨
values=[190,187,184] # 키
colors =[ 'r','g','b'] # 색상

plt.bar(labels,values,color=colors, alpha=0.5) # 막대그래프
```

```python
labels=['강백호','서태웅','정대만'] # 라벨
values=[190,187,184] # 키
plt.bar(labels,values) # 막대그래프
plt.ylim(180,195) # y축 범위
```

```python
labels=['강백호','서태웅','정대만'] # 라벨
values=[190,187,184] # 키
plt.bar(labels,values) # 막대그래프
plt.ylim(180,195) # y축 범위
```

```python
plt.bar(labels,values, width=0.5) # 막대그래프 너비
```

```python
plt.bar(labels,values, width=0.3) # 막대그래프 너비 0.3
plt.xticks(rotation=45) # x축 라벨 회전
plt.yticks(rotation=45) # y축 라벨 회전
```

```python
labels=['강백호','서태웅','정대만'] # 라벨
values=[190,187,184] # 키
ticks =['1번 학생','2번 학생','3번 학생'] # 라벨

plt.bar(labels,values, width=0.5) # 막대그래프
plt.xticks(labels,ticks, rotation = 45) # 라벨 설정
```
