# 2. 출생아 수 및 합계출산율

```python
import pandas as pd
df = pd.read_excel ('142801_20240117144400003_excel.xlsx', skiprows=2, nrows=2, index_col=0)
df
```

```python
df.rename(index={'출생아\xa0수':'출생아 수', '합계\xa0출산율': '합계 출산율'} , inplace=True)
```

```python
df = df.T
df
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='MalGun Gothic'
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus']=False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
plt.plot(df.index, df['출생아 수'])
plt.plot(df.index, df['합계 출산율'])
```

```python
fig, ax1 = plt.subplots(figsize=(15,10)) # 그래프 크기 지정
ax1.plot(df.index, df['출생아 수'], color='blue', marker='o') # 출생아 수 그래프
ax2 = ax1.twinx() # x축을 공유하는 새로운 axes 생성
ax2.plot(df.index, df['합계 출산율'], color='red', marker='o') # 출산율 그래프
```

```python
fig, ax1 = plt.subplots(figsize=(12,5)) # 그래프 크기 지정

ax1.set_ylabel('출생아 수') # 출생아 수 y축 라벨
ax1.set_ylim(250,500)
ax1.set_yticks([300,400,500,600])
ax1.bar(df.index, df['출생아 수'], color='orange') # 출생아 수 그래프


ax2 = ax1.twinx() # x축을 공유하는 새로운 axes 생성
ax2.set_ylabel('합계 출산율') # 출산율 y축 라벨
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1.0])
ax2.plot(df.index, df['합계 출산율'], color='yellow', marker='o', ) # 출산율 그래프
```

```python
fig, ax1 = plt.subplots(figsize=(12,5)) # 그래프 크기 지정

ax1.set_ylabel('출생아 수') # 출생아 수 y축 라벨
ax1.set_ylim(250,500)
ax1.set_yticks([300,400,500,600])
ax1.bar(df.index, df['출생아 수'], color='orange') # 출생아 수 그래프
for idx, val in enumerate (df['출생아 수']):
    ax1.text(idx, val+12, val,  ha='center', fontsize=14) # text(x좌표, y좌표, 텍스트)

ax2 = ax1.twinx() # x축을 공유하는 새로운 axes 생성
ax2.set_ylabel('합계 출산율') # 출산율 y축 라벨
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1.0])
ax2.plot(df.index, df['합계 출산율'], color='yellow', marker='o', ) # 출산율 그래프
```

```python
fig, ax1 = plt.subplots(figsize=(12,5)) # 그래프 크기 지정
fig.suptitle('출생아 수와 합계 출산율', fontsize=20) # 그래프 제목

ax1.set_ylabel('출생아 수') # 출생아 수 y축 라벨
ax1.set_ylim(250,500)
ax1.set_yticks([300,400,500,600])
ax1.bar(df.index, df['출생아 수'], color='orange') # 출생아 수 그래프
for idx, val in enumerate (df['출생아 수']):
    ax1.text(idx, val+12, val,  ha='center', fontsize=14) # text(x좌표, y좌표, 텍스트)

ax2 = ax1.twinx() # x축을 공유하는 새로운 axes 생성
ax2.set_ylabel('합계 출산율') # 출산율 y축 라벨
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1.0])
ax2.plot(df.index, df['합계 출산율'], color='#fcdf03', marker='o', markersize= 12, mec='white', lw=5, mew=3) # 출산율 그래프
for idx, val in enumerate (df['합계 출산율']):
    ax2.text(idx, val+0.05, val,  ha='center', fontsize=14) # text(x좌표, y좌표, 텍스트)

```
