```python
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='MalGun Gothic'
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus']=False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python

data = {
    '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
    '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
    '관객 수(만명)' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
    '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)
df
```

```python
plt.bar(df['영화'], df['평점'])
plt.show()
```

```python
plt.bar(df['영화'], df['평점'])
plt.xlabel('영화')
plt.ylabel('평점')
plt.xticks(rotation=90)
plt.title('국내 Top 8 영화 평점 정보')
plt.show()
```

```python
df_group =df.groupby('개봉 연도').mean(numeric_only=True)
df_group
```

```python
import numpy as np
plt.plot(df_group.index.astype(int), df_group['평점'], marker='o', markersize=10, color='red', linestyle='--', label='평점')
plt.xticks(list(range(2005, 2021, 5)))
plt.show()
print(df_group.index.astype(int))
```

```python
df_group.plot(kind='line', y='평점', marker='o', color='b',  figsize=(10, 5))
plt.show()
```

```python
df['Sorted'] = df['평점'].apply(lambda x: '9점 이상' if x > 9 and x <= 9.5 else '9점 미만')
print(df)
df_group_score = df.groupby('Sorted').count()
df_group_score
df_group_score.plot(kind='pie', y='영화', autopct='%.2f%%', figsize=(5, 5))
plt.show()
```

```python
import matplotlib.pyplot as plt

sizes = [215, 130, 245, 210]
labels = ['Python', 'C++', 'Ruby', 'Java']
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
