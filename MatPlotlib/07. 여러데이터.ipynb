## COVID-19 백신 종류별 접종 연구

```python
days =[1,2,3] # 1일, 2일, 3일
az = [2,4,8] # (단위: 만명) 1일부터 3일까지 아스트라제네카 백신 접종자 수
pfizer = [5,1,3] # 화이자 백신 접종자 수
moderna = [1,2,5] # 모더나 백신 접종자 수

plt.plot(days, az, label='아스트라제네카')
plt.plot(days, pfizer, label='화이자', marker='o', linestyle='--')
plt.plot(days, moderna, label='모더나', marker='s', linestyle='-.')
plt.legend()
```

```python
plt.plot(days, az, label='아스트라제네카')
plt.plot(days, pfizer, label='화이자', marker='o', linestyle='--')
plt.plot(days, moderna, label='모더나', marker='s', linestyle='-.')
plt.legend(ncol=3, loc=2) # 범례를 3열로 표시
```
